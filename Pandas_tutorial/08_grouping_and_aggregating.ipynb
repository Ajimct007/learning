## .gropuby Function

## Pandas documentation says a groupby operation involves some combination of Splitting the object, applying a function and  combining results

```python
# for reference, first look at the value_counts for countries
df['Country'].value_counts()
```

```python
# How to use the groupby function?
# first splitting the object

df.groupby(['Country'])

#by Running this we got "DataFrameGroupBy object"

#this object contains bunch of groups
```

```python
#giving a variable for the above

country_grp = df.groupby(['Country'])
```

```python
#access based on the variable
country_grp
```

```python
#grabbing the specific group by country name

country_grp.get_group('United States')

# we got the results only for the Country name with United States
```

```python
# It is similar to the filter method which we have used previous parts

filt = df['Country'] == 'United States'
df.loc[filt]

#it shows the same as the above but it show only for the one country
```

```python
# groupby splits the all the values in the country column
```

### since we split up using the groupby, now we can apply a function to bring the results together

```python
## Question --Most populat Social media platforms for country
# Method 1 -- Without groupby method or using the filter method

filt = df['Country'] == 'United States'
df.loc[filt]['SocialMedia'].value_counts()

#It shows the results for United states
```

## We can use this to the Aggregate functions like "Mean, Median and Mode"

```python
#for seeing median of the salary column
country_grp['ConvertedComp'].median() #shows the results for all the countries
```

```python
# For getting the results based on a single country
country_grp['ConvertedComp'].median().loc['Germany']
```

```python
# For getting multiple aggregate functions
country_grp['ConvertedComp'].agg(['median','mean']).loc['Canada']
```

```python
# Using the filtering Method to get Language(Python) based on the county 

filt = df['Country'] == 'India'
df.loc[filt]['LanguageWorkedWith'].str.contains('Python')

# It returns the values of True or false
```

```python
# for getting how many people knows Python, we can use .sum(), it works with boolean and assumes True=1, False=0
filt = df['Country'] == 'India'
df.loc[filt]['LanguageWorkedWith'].str.contains('Python').sum()
```

```python
# using .sum() to groupby object method

country_grp['LanguageWorkedWith'].str.contains('Python').sum()

# by running this we got an error

# We cannot access the attribute str for SeriesGropuBy object so we have to use the apply method
```

```python
#here we are using the apply methos for sum up and see the results using the lambda function

country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())

# It shows the results for entire series of country
```

## Question -- What % of people from each country know Python?

```python
# First get the total number of respondents for each country

country_respondants = df['Country'].value_counts()
country_respondants
```

```python
country_uses_python = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_uses_python
```

```python
## We can combine more than one seried using the Pandas concat function
## we are creating a new dataframe

python_df = pd.concat([country_respondants,country_uses_python], axis='columns',sort=False)
python_df

# created the df
```

```python
# we are renaming the column names
python_df.rename(columns={'Country':'NumRespondants','LanguageWorkedWith':'NumKnowsPython'},inplace=True)
```

```python
python_df
```

```python
## Create a new column to get the percentage value
python_df['PctKnowsPython'] = (python_df['NumKnowsPython']/python_df['NumRespondants'])*100
```

```python
python_df
```

```python
## Sort the country based on largest % of Respondents  know Pyhton
python_df.sort_values(by='PctKnowsPython', ascending=False, inplace=True)
```
