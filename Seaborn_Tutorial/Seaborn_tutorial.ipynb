```python
## Importing the important libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# total bill vs dataset
tips = sns.load_dataset('tips')
tips.head()
```

```python
# setting a theme for the plots
sns.set_theme()
```

```python
import collections
# visualize the data
sns.relplot(data=tips, x ='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
plt.title('Bills vs Tips')
```

```python
## load iris dataset
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
## Scatter plot
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=iris)
```

```python
sns.scatterplot(x='sepal_length', y='petal_width', hue='species', data=iris)
```

```python
# load titanic dataset
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
## Count Plot
sns.countplot(x='class', data=titanic)
```

```python
sns.countplot(x='survived',data=titanic)
```

```python
#Barplot
sns.barplot(x='sex',y='survived',hue='class',data=titanic)
```

```python
corr = titanic.corr()
```

```python
## constructing a heatmap
plt.figure(figsize=(10,10))
sns.heatmap(corr, cbar=True, square=False, fmt='.1f', annot=True, annot_kws={'size':15}, cmap='Blues')
```
