# Updating rows and columns -- Modifying Data with DataFrames

```python
import pandas as pd
```

```python
people = {
    "first": ["Corey", 'Jane', 'John'], 
    "last": ["Schafer", 'Doe', 'Doe'], 
    "email": ["CoreyMSchafer@gmail.com", 'JaneDoe@email.com', 'JohnDoe@email.com']
}
```

```python
df = pd.DataFrame(people)
```

```python
df
```

## Updating Columns


```python
df.columns
```

```python
# for changing the column name 'last' into 'last_name' and 'first' into 'first_name' we can use the columns attribute
df.columns = ['first_name', 'last_name', 'email']
```

```python
df.columns # we can observe the updated columns here
```

```python
df
```

```python
# updating the column names with upper case
df.columns = [x.upper() for x in df.columns]
```

```python
df # Here we can observe the Uppercase letters
```

```python
#for changing the '_' into ''
df.columns = df.columns.str.replace('_',' ')
```

```python
df
```

```python
#here we have created the space now we want back the '_'
df.columns = df.columns.str.replace(' ','_')
```

```python
df
```

```python
#for changing the letters to lower case again
df.columns = [x.lower() for x in df.columns]
df
```

### There are 4 more methods to change the row values we will go one by one
#### The 4 methods are
#### 1) apply
#### 2) map
#### 3) applymap
#### 4) replace


### First go with apply

```python
# apply is used for calling a function on our values -- it can work on either a dataframe or a series object


# lets look how apply work for a series?
df['email'].apply(len) #it applies the len() function to the email column and gives the value of counts in email
```

```python
# Writing a function to return uppercase
def update_email(email):
    return email.upper()
```

```python
#here we are calling the update_email function using the apply
df['email'].apply(update_email) #we can see the email values in upper case and it not chnage in the df
```

```python
df
```

```python
df['email'] = df['email'].apply(update_email)
```

```python
df #now the email changed to uppercase
```

```python
#using the lambda function to convert the email to lowercase
df['email'] = df['email'].apply(lambda x: x.lower())
```

```python
df #converted to lowercase
```

## We are going to see how apply works with dataframes

```python
df['email'].apply(len)
```

```python
df.apply(len) #not applying the len function to dataframe, but applyng to each series of the dataframe, specifically the columns
#or the number of columns
```

```python
#manually checking the columns
len(df['email'])#it gives us 3
```

```python
#we can apply this to rows as well
df.apply(len,axis='columns') #it counts the columns or count from left to right or vice versa 
```

```python
# for getting the minimum value from each column, series object has min method
df.apply(pd.Series.min) # it provides the result in alphabetically arranged order for each column
```

```python
# We can also work with lambda function as well
# Remember Lambda will be working with series object
df.apply(lambda x: x.min()) # it gives the same response as above
```

This is more useful when the dataset contains numerical data, eg: for applying the numpy squareroot for numerical objects

## Apply_Map

### for applying a function to every individual element in the dataframe that's where Apply_map is used for!!!
### Apply_map only works on dataframes

```python
df.applymap(len) #it applies the len function to each individual value in our dataframe
```

```python
#since our dataframe contain strings we can convert them to lowercase using the applymap
df.applymap(str.lower)
# if we had numerical data in our dataframe, we will get an error
```

## MAP
### It only works on a series
### used for subsituting each value in a series with an other value

```python
# Let's say we want to subsitute the first name with other names
df['first'].map({'Corey': 'Chris', 'Jane':'Mary'})
# The value that we didn't converted changed to NaN 
```

```python
# Since it turns the non mentioned values to NaN we have to use the replace method
```

## Replace

```python
df['first'].replace({'Corey':'Chris','Jane':'Mary'})
# By executing this it will not change the un-mentioned value to NaN

# But it not changed in the dataframe for that to occur
```

```python
# We have to specify the column
df['first'] = df['first'].replace({'Corey':'Chris','Jane':'Mary'})
```
