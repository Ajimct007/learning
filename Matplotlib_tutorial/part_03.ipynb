# Pie charts

```python
from matplotlib import pyplot as plt
plt.style.use("fivethirtyeight")
```

```python
slices = [120, 80]
labels = ['Sixty', 'Forty']
```

```python
plt.pie(slices, labels=labels)
plt.title("Simpl Pie chart")
plt.tight_layout()
```

```python
# Adding wedge properties for Highlighting
plt.pie(slices, labels=labels, wedgeprops={'edgecolor':'black'}) #we can search in google for wedge properties
plt.title("Simpl Pie chart")
plt.tight_layout()
```

```python
slices = [120, 80, 30, 20]
labels = ['Sixty', 'Forty', 'Extra1', 'Extra2']
colours = ['blue', 'red', 'yellow', 'green']
```

```python
#Passing the colors


plt.pie(slices, labels=labels,colors=colours , wedgeprops={'edgecolor':'black'})
plt.title("Simpl Pie chart")
plt.tight_layout()
```

```python
# We can use the hex colours
slices = [120, 80, 30, 20]
labels = ['Sixty', 'Forty', 'Extra1', 'Extra2']
colours = ['#008fd5', '#fc4f30', '#e5ae37', '#6d904f']
```

```python
#Passing hex colors


plt.pie(slices, labels=labels,colors=colours , wedgeprops={'edgecolor':'black'})
plt.title("Simpl Pie chart")
plt.tight_layout()
```

```python
#Adding explode

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0,0,0,0.1,0]
```

```python
plt.pie(slices, labels=labels, explode=explode, wedgeprops={'edgecolor':'black'})
plt.title("Simpl Pie chart")
plt.tight_layout()
```

```python
# Addign shadow, starting angle, for displaying percentage

plt.pie(slices, labels=labels, explode=explode, shadow=True, 
        startangle=90, autopct='%1.1f%%', 
        wedgeprops={'edgecolor':'black'})
plt.title("Simpl Pie chart")
plt.tight_layout()
```
