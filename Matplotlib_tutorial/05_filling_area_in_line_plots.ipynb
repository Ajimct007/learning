# Filling Area on Line Plots

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('D:\Data Science\Matplotlib_tutorial\part_05\data.csv')
```

```python
data.head()

# The dataset have ages and salaries with respect to prog languages
```

```python
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']
```

```python
# simple line graph

plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()
```

```python
### adding fills

plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

## adding fill

plt.fill_between(ages, py_salaries)

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()
```

```python
plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

## adding fill and alpha

plt.fill_between(ages, py_salaries, alpha=0.25)

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()
```

```python
## using conditional thresholds

plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

## adding fill and alpha

## we have the median salary for overall developers which is pre calculated

overall_median = 57287

#adding Y2 value

plt.fill_between(ages, py_salaries, overall_median, alpha=0.25)

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

## since we set the threshold for python, it fills the shade above the line till the threshold and fills below after above the salary
```

```python
## we can put some additional conditions, it only fills only when the condition is met


plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')


overall_median = 57287

#adding a condition here

plt.fill_between(ages, py_salaries, overall_median,
                 where=(py_salaries > overall_median), 
                 interpolate=True, alpha=0.25)

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

# we can see the it filled only where our median salary is filled above our conditinal salary of median value
```

```python
## we can put some additional conditions, it only fills only when the condition is met


plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')


overall_median = 57287

#adding a condition here

plt.fill_between(ages, py_salaries, overall_median,
                 where=(py_salaries > overall_median), 
                 interpolate=True, alpha=0.25)

# for plotting area below with different colour

plt.fill_between(ages, py_salaries, overall_median,
                 where=(py_salaries <= overall_median), 
                 interpolate=True, color = 'red' ,alpha=0.25)

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

```

```python
## we can put some additional conditions, it only fills only when the condition is met
## fill between 2 lines

plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')


overall_median = 57287
#insted of using the overall_mean we will use the dev salaries

#adding a condition here

plt.fill_between(ages, py_salaries, dev_salaries,
                 where=(py_salaries > dev_salaries), 
                 interpolate=True, alpha=0.25, label='Above Avg')

# for plotting area below with different colour

plt.fill_between(ages, py_salaries, dev_salaries,
                 where=(py_salaries <= dev_salaries), 
                 interpolate=True, color = 'red' ,alpha=0.25, label='Below Avg')

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()


## Here we can see it fills between the salaries
```
