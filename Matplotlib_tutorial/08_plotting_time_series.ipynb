# Plotting Time Series Data

```python
import pandas as pd
from datetime import datetime, timedelta
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates
```

```python
plt.style.use('seaborn')
```

```python
dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]
y = [0, 1, 3, 4, 6, 5, 7]
```

```python
# plotting the dates

plt.plot_date(dates, y)


plt.tight_layout()

plt.show()

## It plots the data points in a scatter plot
```

```python
## making a line

plt.plot_date(dates, y, linestyle='solid')


plt.tight_layout()

plt.show()
```

```python
## adding format

plt.plot_date(dates, y, linestyle='solid')

# code for auto format
plt.gcf().autofmt_xdate()

plt.tight_layout()

plt.show()
```

## link for date formatting -- https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavior


```python
# changing the format
# date format of month, day year

plt.plot_date(dates, y, linestyle='solid')

plt.gcf().autofmt_xdate() ## gct means get current format

date_format = mpl_dates.DateFormatter('%b, %d, %Y')

plt.gca().xaxis.set_major_formatter(date_format) # gca means get current axis

# now it shows the month in the plot!!!

plt.tight_layout()

plt.show()
```

# Applying to real data

```python
data = pd.read_csv('D:\\Data Science\\Matplotlib_tutorial\\part_08\\data.csv')
```

```python
price_date = data['Date']
price_close = data['Close']
```

```python
## Passing the price date and price close to the graph

plt.plot_date(price_date, price_close, linestyle='solid')

plt.gcf().autofmt_xdate()

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')

plt.tight_layout()

plt.show()
```

```python
## our date is not in date series, so we have to convert it
data['Date'] = pd.to_datetime(data['Date'])
```

```python
#sorting the data
data.sort_values('Date', inplace=True)
```

```python
plt.plot_date(price_date, price_close, linestyle='solid')

plt.gcf().autofmt_xdate()

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')

plt.tight_layout()

plt.show()
```
