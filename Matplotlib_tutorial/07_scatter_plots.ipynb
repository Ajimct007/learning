# Scatter plots

```python
import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('seaborn')
```

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
```

```python
# Creating a basic scatter plot

plt.scatter(x,y)
plt.tight_layout()
```

```python
# changing the size of the dots, colour


plt.scatter(x,y, s = 100, c='green', marker='X') 

# s is for size
# C is for colour

plt.tight_layout()
```

```python
# adding edge colour

plt.scatter(x,y, s = 100, c='green', edgecolor='black', linewidth=1, alpha=0.5) 

# edge colour -- colouring the edge of data-point
# linewidth -- giving the width to the edge of data-point
# alpha -- for providing contrast to data-point

# s is for size
# C is for colour

plt.tight_layout()
```

```python
## We have list of Colours which assumed as rating from 1 to 10

colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

## passing the list of colors to the c attribute in the plt

plt.scatter(x,y, s = 100, c=colors, edgecolor='black', linewidth=1, alpha=0.75) 

# edge colour -- colouring the edge of data-point
# linewidth -- giving the width to the edge of data-point
# alpha -- for providing contrast to data-point

# s is for size
# C is for colour

plt.tight_layout()


#We can observe the different colours based of rating, currently we are getting the grey we can change that
```

```python
# set colour map or cmap as Greens

plt.scatter(x,y, s = 100, c=colors, cmap = 'Greens' , edgecolor='black', linewidth=1, alpha=0.75) 

## Adding legend

cbar = plt.colorbar()

cbar.set_label('Satisfaction')


plt.tight_layout()
```

```python
#We have the list of sizes which we can pass to our plot

sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174,
         538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]

#adding sizes to the s 

plt.scatter(x,y, s = sizes, c=colors, cmap = 'Greens' , edgecolor='black', linewidth=1, alpha=0.75) 

## Adding legend

cbar = plt.colorbar()

cbar.set_label('Satisfaction')


plt.tight_layout()
```

# Applying the learning to real world data

```python
# Importing youtube data, top 200 trending videos with views and likes and the ratio

data = pd.read_csv('D:\\Data Science\\Matplotlib_tutorial\\part_07\\2019-05-31-data.csv')
```

```python
data.head()
```

```python
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']
```

```python


plt.scatter(view_count,likes,
           edgecolors='black', linewidth=1, alpha=0.75)
plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Views')
plt.tight_layout()
```

```python
# Using log scale 

plt.scatter(view_count,likes,
           edgecolors='black', linewidth=1, alpha=0.75)

## Using the log scale for x and y axis

plt.xscale('log')
plt.yscale('log')


plt.title('Trending YouTube Videos')

plt.xlabel('View Count')

plt.ylabel('Total Views')
plt.tight_layout()
```

```python
# using the ratio of likes and dislikes

plt.scatter(view_count,likes, c=ratio, cmap = 'summer',
           edgecolors='black', linewidth=1, alpha=0.75)

cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')

## Using the log scale for x and y axis

plt.xscale('log')
plt.yscale('log')


plt.title('Trending YouTube Videos')

plt.xlabel('View Count')

plt.ylabel('Total Views')
plt.tight_layout()
```
