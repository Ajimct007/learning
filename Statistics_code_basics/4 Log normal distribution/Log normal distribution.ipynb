```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("D:\\Data Science\\Code basics\\math-for-machine-learning-main\\5_log_normal_distribution\\income.csv", 
    index_col=None, 
    names=["income","count"], 
    skiprows=1
)
df.head()
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
g = sns.barplot(x='income',y='count',data=df)
g.set_xticklabels(g.get_xticklabels(), 
                          rotation=45, 
                          horizontalalignment='right');
```
