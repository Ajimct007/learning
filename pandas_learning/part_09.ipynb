# Cleaning Data - Casting Datatypes and Handling Missing Values

```python
import pandas as pd
import numpy as np
```

```python
# Python dictionary with key values
people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}

### Here we have multiple null values
# 1) np.nan -- NaN
# 2) None
# 3) NA
# 4) Missing
```

```python
#loading as Dataframe
df = pd.DataFrame(people)
```

```python
df
```

```python
df.dropna() #when we run this without any argument, we get 4 rows insted of 7
#it using the default arguments
```

```python
df.dropna(axis='index', how='any')#these are the default arguments
```

```python
df.dropna(axis='index', how='all')
# axis can be set to index or columns
# how -- 'any' : If any NA values are present, drop that row or column.
# how -- 'all' : If all values are NA, drop that row or column.
```

```python
df.dropna(axis='columns', how='any') #since all columns has atleast 1 NaN value it removed them
```

```python
df
```

```python
#for dropping NaN at specific rows of a column
df.dropna(axis='index', how='any', subset=['email'])
```

```python
# We can also pass multiple columns to our subset
# to make the data to be useful then we need either lastname or their email address but we dont need both

df.dropna(axis='index', how='all', subset=['last','email'])

# it looks for both the columns of last and email, if any of the column is filled then it won't drop it
# to make it reflect in the df we have to give inplace=True
```

```python
## how to handle the custom missing values
## i.e -- NA, Missing
# it dependes on the dataset

# we can use the replace function
df.replace('NA',np.nan, inplace=True) #replacing the NA as NaN
df.replace('Missing',np.nan, inplace=True)
```

```python
df
```

```python
## if we run the dropna, it will remove the old and changed one
df.dropna()
```

```python
df.isna() #shows the mask of values which has na or not
```

```python
# for filling the Na values
df.fillna(0)
```

### casting datatypes

```python
df.dtypes # it show the columns as object
# object means it is a string, a number or mix of both
```

```python
# Since age column is object, we cant see the object
df['age'].mean()
```

```python
# The error says that the columns are strings not integers

# When we have the NaN values in a column that we trying to convert to numbers then we need to use the float datatype since

# Since the NaN value is actually a float

type(np.nan) #it shows the nan as float
```

```python
df['age'] = df['age'].astype(int) # Since the age column has some nan value we cant convert it to the integer
# but we can convert them to float so that NaN value stay missing
```

```python
# converting the age column into float
df['age'] = df['age'].astype(float)
```

```python
df.dtypes #now the age column is float
```

```python
# Now we take the average of age
df['age'].mean()
```

## Casting values
## Question - For the developers who answered the survey we want to calculate the average years of coding experience
## Why it is so difficult to calculate for this dataset

```python
# The column for years of experience is calles Years code
df['YearsCode'].head(10)
```

```python
df['YearsCode'].mean()

#it shows error -- can only concatenate str (not "int") to str

# So we have to convert it into float
```

```python
# Convert the YearsCode to float

df['YearsCode'] = df['YearsCode'].astype(float)

# Even after converting it to float, we are still getting error

# Error -- could not convert string to float: 'Less than 1 year'

# There is a String value of 'Less than 1 year'
```

```python
# Let's see unique value in this column

df['YearsCode'].unique()

# We have the strings of 'Less than 1 year', 'More than 50 years', we have to replace them into numbers
```

```python
df['YearsCode'].replace('Less than 1 year', 0 , inplace=True) #replacing the Less than 1 year to 0
```

```python
df['YearsCode'].replace('More than 50 years', 51 , inplace=True) 
```

```python
df['YearsCode'].unique() #Now we changed the string values
```

```python
# Now convert it to float

df['YearsCode'] = df['YearsCode'].astype(float)

# Since we converted the strings to integer we are not facing any error
```

```python
df['YearsCode'].dtypes #here we found that datatype changed to float64
```

```python
# Now we can see the average years of coding
df['YearsCode'].mean()
```
